## 🌐 Bloque 1: Protocolo HTTP y Consumo de APIs

### 📘 Teoría

**HTTP (HyperText Transfer Protocol)** es el protocolo que define cómo se comunican cliente y servidor en la web.

#### Flujo de una solicitud HTTP
```
  CLIENTE                          SERVIDOR
    │                                  │
    │  ── GET /usuarios/1 ──────────►  │
    │                                  │  busca usuario #1
    │  ◄── 200 OK + JSON ───────────   │
    │                                  │
    │  ── POST /usuarios ───────────►  │
    │     { nombre: 'Ana' }            │  crea nuevo usuario
    │  ◄── 201 Created ─────────────   │
```

#### Métodos HTTP

| Método | Acción | Ejemplo |
|--------|--------|---------|
| `GET` | Leer / obtener datos | `GET /productos/5` |
| `POST` | Crear un recurso nuevo | `POST /productos` |
| `PUT` | Reemplazar un recurso completo | `PUT /productos/5` |
| `PATCH` | Modificar parcialmente | `PATCH /productos/5` |
| `DELETE` | Eliminar | `DELETE /productos/5` |

#### Códigos de estado más importantes

| Código | Significado |
|--------|-------------|
| `200 OK` | Todo bien, respuesta exitosa |
| `201 Created` | Recurso creado exitosamente |
| `400 Bad Request` | El cliente mandó datos inválidos |
| `401 Unauthorized` | Falta autenticación |
| `403 Forbidden` | Autenticado pero sin permiso |
| `404 Not Found` | Recurso no existe |
| `500 Internal Server Error` | Error en el servidor |

#### Estructura de una URL de API REST
```
https://api.ejemplo.com/v1/productos?categoria=electronica&limite=10
│─────────────────────│ │──────────│ │──────────────────────────────│
       base URL          endpoint          query parameters
```

### 💡 Ejemplo resuelto 1 — Consumir una API pública con `requests`

````python
import requests
import json

# ✅ Consumir la API pública JSONPlaceholder (no requiere clave)
# https://jsonplaceholder.typicode.com — API de prueba gratuita

BASE_URL = 'https://jsonplaceholder.typicode.com'

def get_usuario(user_id):
    """Obtiene un usuario por ID."""
    response = requests.get(f'{BASE_URL}/users/{user_id}')
    print(f'GET /users/{user_id} → Status: {response.status_code}')
    if response.status_code == 200:
        return response.json()
    return None

def get_posts_usuario(user_id):
    """Obtiene los posts de un usuario."""
    response = requests.get(f'{BASE_URL}/posts', params={'userId': user_id})
    print(f'GET /posts?userId={user_id} → Status: {response.status_code}')
    return response.json() if response.status_code == 200 else []

def crear_post(titulo, cuerpo, user_id):
    """Simula crear un post (JSONPlaceholder no persiste, pero responde 201)."""
    payload = {'title': titulo, 'body': cuerpo, 'userId': user_id}
    response = requests.post(f'{BASE_URL}/posts', json=payload)
    print(f'POST /posts → Status: {response.status_code}')
    return response.json()

# ─── Ejecutar ───
print('=== Obtener usuario ===')
usuario = get_usuario(1)
if usuario:
    print(f'  Nombre:   {usuario["name"]}')
    print(f'  Email:    {usuario["email"]}')
    print(f'  Ciudad:   {usuario["address"]["city"]}')
    print(f'  Empresa:  {usuario["company"]["name"]}')

print('\n=== Posts del usuario 1 ===')
posts = get_posts_usuario(1)
for p in posts[:3]:  # primeros 3
    print(f'  [{p["id"]}] {p["title"][:50]}...')
print(f'  ... ({len(posts)} posts en total)')

print('\n=== Crear un nuevo post ===')
nuevo = crear_post('Mi primer post', 'Contenido del post de prueba', 1)
print(f'  Post creado con ID: {nuevo["id"]}')
print(f'  Título: {nuevo["title"]}')

# Headers de la respuesta
resp = requests.get(f'{BASE_URL}/users/1')
print('\n=== Headers de respuesta ===')
print(f'  Content-Type: {resp.headers.get("Content-Type")}')
print(f'  Tamaño:       {resp.headers.get("Content-Length", "N/A")} bytes')
````

## 🖥️ Bloque 2: Servidor Web con Flask

### 📘 Teoría

**Flask** es un microframework web para Python. Te permite crear un servidor HTTP con muy poco código.

#### Conceptos clave

```python
from flask import Flask, request, jsonify

app = Flask(__name__)   # crear la aplicación

# Una 'ruta' asocia una URL a una función Python
@app.route('/saludo', methods=['GET'])
def saludo():
    nombre = request.args.get('nombre', 'mundo')  # query param
    return jsonify({'mensaje': f'Hola, {nombre}!'})

@app.route('/usuarios', methods=['POST'])
def crear_usuario():
    datos = request.get_json()   # leer body JSON
    # procesar...
    return jsonify({'id': 1, **datos}), 201  # 201 = Created

if __name__ == '__main__':
    app.run(debug=True)   # corre en http://localhost:5000
```

#### Estructura de una API REST con Flask

| URL | Método | Acción |
|-----|--------|--------|
| `/productos` | GET | Listar todos |
| `/productos/<id>` | GET | Obtener uno |
| `/productos` | POST | Crear nuevo |
| `/productos/<id>` | PUT | Actualizar |
| `/productos/<id>` | DELETE | Eliminar |

> 💡 **Nota:** En este notebook probamos Flask usando el cliente de prueba integrado (`app.test_client()`)
> para poder ejecutar todo dentro del notebook sin necesitar un servidor externo.

### 💡 Ejemplo resuelto 2 — API REST de productos con Flask

````python
from flask import Flask, request, jsonify
import json

# ✅ API REST completa para gestión de productos

app = Flask(__name__)

# Base de datos en memoria (en producción sería SQLite/PostgreSQL)
productos_db = [
    {'id': 1, 'nombre': 'Laptop Pro',  'precio': 1200.00, 'stock': 15, 'categoria': 'Electrónica'},
    {'id': 2, 'nombre': 'Auriculares', 'precio':   89.99, 'stock': 50, 'categoria': 'Electrónica'},
    {'id': 3, 'nombre': 'Camiseta',    'precio':   25.00, 'stock':100, 'categoria': 'Ropa'},
]
next_id = 4

@app.route('/productos', methods=['GET'])
def listar_productos():
    categoria = request.args.get('categoria')
    if categoria:
        filtrados = [p for p in productos_db if p['categoria'] == categoria]
        return jsonify({'productos': filtrados, 'total': len(filtrados)})
    return jsonify({'productos': productos_db, 'total': len(productos_db)})

@app.route('/productos/<int:prod_id>', methods=['GET'])
def obtener_producto(prod_id):
    prod = next((p for p in productos_db if p['id'] == prod_id), None)
    if prod is None:
        return jsonify({'error': f'Producto {prod_id} no encontrado'}), 404
    return jsonify(prod)

@app.route('/productos', methods=['POST'])
def crear_producto():
    global next_id
    datos = request.get_json()
    if not datos or 'nombre' not in datos or 'precio' not in datos:
        return jsonify({'error': 'Se requieren nombre y precio'}), 400
    nuevo = {'id': next_id, 'nombre': datos['nombre'],
             'precio': datos['precio'], 'stock': datos.get('stock', 0),
             'categoria': datos.get('categoria', 'Sin categoría')}
    productos_db.append(nuevo)
    next_id += 1
    return jsonify(nuevo), 201

@app.route('/productos/<int:prod_id>', methods=['PUT'])
def actualizar_producto(prod_id):
    prod = next((p for p in productos_db if p['id'] == prod_id), None)
    if prod is None:
        return jsonify({'error': 'No encontrado'}), 404
    datos = request.get_json()
    prod.update(datos)
    return jsonify(prod)

@app.route('/productos/<int:prod_id>', methods=['DELETE'])
def eliminar_producto(prod_id):
    global productos_db
    prod = next((p for p in productos_db if p['id'] == prod_id), None)
    if prod is None:
        return jsonify({'error': 'No encontrado'}), 404
    productos_db = [p for p in productos_db if p['id'] != prod_id]
    return jsonify({'mensaje': f'Producto {prod_id} eliminado'}), 200


# ── Probar con el cliente de prueba de Flask ──
with app.test_client() as client:

    # GET todos
    r = client.get('/productos')
    data = r.get_json()
    print(f'GET /productos → {r.status_code}')
    print(f'  Total: {data["total"]} productos')

    # GET por categoría
    r = client.get('/productos?categoria=Electrónica')
    data = r.get_json()
    print(f'\nGET /productos?categoria=Electrónica → {r.status_code}')
    for p in data['productos']:
        print(f'  {p["nombre"]:15} | ${p["precio"]:.2f}')

    # GET por ID
    r = client.get('/productos/1')
    print(f'\nGET /productos/1 → {r.status_code}')
    print(f'  {r.get_json()}')

    # GET no existente
    r = client.get('/productos/999')
    print(f'\nGET /productos/999 → {r.status_code} (esperado 404)')
    print(f'  {r.get_json()}')

    # POST crear
    r = client.post('/productos',
                    data=json.dumps({'nombre': 'Monitor 4K', 'precio': 450.00, 'categoria': 'Electrónica'}),
                    content_type='application/json')
    print(f'\nPOST /productos → {r.status_code} (esperado 201)')
    print(f'  Creado: {r.get_json()}')

    # PUT actualizar
    r = client.put('/productos/1',
                   data=json.dumps({'precio': 999.99}),
                   content_type='application/json')
    print(f'\nPUT /productos/1 → {r.status_code}')
    print(f'  Actualizado: {r.get_json()}')

    # DELETE
    r = client.delete('/productos/2')
    print(f'\nDELETE /productos/2 → {r.status_code}')
    print(f'  {r.get_json()}')
````

### 💡 Ejemplo resuelto 3 — Análisis de empleados con Pandas

````python
import pandas as pd
import numpy as np

# ✅ Crear un DataFrame realista de empleados

np.random.seed(42)
n = 50

ciudades   = ['Buenos Aires', 'Córdoba', 'Rosario', 'Mendoza', 'Tucumán']
deptos     = ['IT', 'Ventas', 'RRHH', 'Marketing', 'Finanzas']
niveles    = ['Junior', 'Semi-Senior', 'Senior']

df = pd.DataFrame({
    'nombre':     [f'Empleado_{i:02d}' for i in range(n)],
    'edad':       np.random.randint(22, 58, n),
    'ciudad':     np.random.choice(ciudades, n),
    'depto':      np.random.choice(deptos, n),
    'nivel':      np.random.choice(niveles, n),
    'salario':    np.random.randint(40_000, 150_000, n).astype(float),
    'años_exp':   np.random.randint(0, 20, n),
    'activo':     np.random.choice([True, False], n, p=[0.9, 0.1]),
})

# Introducir algunos nulos para practicar limpieza
df.loc[np.random.choice(df.index, 5), 'salario'] = np.nan

print('=== Información general ===')
print(f'  Forma: {df.shape[0]} filas × {df.shape[1]} columnas')
print(f'  Columnas: {list(df.columns)}')
print(f'  Valores nulos:\n{df.isnull().sum()[df.isnull().sum() > 0]}')

print('\n=== Primeras 5 filas ===')
print(df.head())

print('\n=== Estadísticas descriptivas (numéricas) ===')
print(df[['edad', 'salario', 'años_exp']].describe().round(2))

# Limpieza: rellenar nulos con la mediana del departamento
df['salario'] = df.groupby('depto')['salario'].transform(lambda x: x.fillna(x.median()))
print(f'\n✅ Nulos en salario después de limpiar: {df["salario"].isnull().sum()}')

# Filtros
senior_it = df[(df['nivel'] == 'Senior') & (df['depto'] == 'IT')]
print(f'\n=== Seniors en IT: {len(senior_it)} empleados ===')
print(senior_it[['nombre', 'ciudad', 'salario', 'años_exp']].to_string(index=False))

# Agrupaciones
print('\n=== Salario promedio por departamento ===')
por_depto = df.groupby('depto')['salario'].agg(['mean', 'min', 'max', 'count'])
por_depto.columns = ['promedio', 'mínimo', 'máximo', 'cantidad']
por_depto['promedio'] = por_depto['promedio'].round(0)
print(por_depto.sort_values('promedio', ascending=False))
````

#### Broadcasting
NumPy puede operar arrays de distintas formas ajustando automáticamente las dimensiones:

```

### 💡 Ejemplo resuelto 4 — NumPy para análisis de datos de sensores

````python
import numpy as np

# ✅ Simular y analizar datos de sensores de temperatura

np.random.seed(42)

# Simular lecturas de 5 sensores durante 30 días (720 horas)
n_sensores = 5
n_horas    = 720  # 30 días

# Cada sensor tiene una temperatura base diferente + ruido
bases = np.array([20.0, 22.5, 18.0, 25.0, 21.0])  # shape (5,)
datos = bases[:, np.newaxis] + np.random.normal(0, 2, (n_sensores, n_horas))
# datos shape: (5 sensores, 720 horas)

print(f'Forma del array de datos: {datos.shape}')
print(f'Dtype: {datos.dtype}')

# ─ Estadísticas por sensor ─
print('\n=== Estadísticas por sensor ===')
print(f'  {"Sensor":8} {"Media":>8} {"Min":>8} {"Max":>8} {"StdDev":>8}')
print('  ' + '-'*44)
for i in range(n_sensores):
    print(f'  Sensor_{i+1}  {datos[i].mean():8.2f} {datos[i].min():8.2f} {datos[i].max():8.2f} {datos[i].std():8.2f}')

# ─ Detección de anomalías (valores fuera de ±2 desvíos) ─
medias = datos.mean(axis=1, keepdims=True)   # media de cada sensor
stds   = datos.std(axis=1, keepdims=True)    # std de cada sensor
anomalias = np.abs(datos - medias) > 2 * stds
print(f'\n=== Anomalías detectadas ===')
for i in range(n_sensores):
    n_anom = anomalias[i].sum()
    pct    = n_anom / n_horas * 100
    print(f'  Sensor_{i+1}: {n_anom:3d} anomalías ({pct:.1f}%)')

# ─ Promedios diarios (reshape para agrupar por día) ─
datos_diarios = datos.reshape(n_sensores, 30, 24).mean(axis=2)
print(f'\n=== Forma después del reshape: {datos_diarios.shape} (sensores, días)')
print('  Temperatura promedio diaria — primeros 5 días, Sensor_1:')
print(f'  {datos_diarios[0, :5].round(2)}')

# ─ Correlación entre sensores ─
print('\n=== Correlación entre sensores ===')
correlacion = np.corrcoef(datos)
print('  Matriz de correlación (redondeada):')
print(np.round(correlacion, 2))
````

### 💡 Ejemplo resuelto 5 — Pipeline completo: API → Pandas → Análisis

````python
import pandas as pd
import numpy as np

# ✅ Generamos un dataset sintético de ventas de e-commerce
# (simula lo que obtendrías de una API real)

np.random.seed(123)
n = 500

fechas     = pd.date_range('2023-01-01', periods=n, freq='D')
categorias = ['Electrónica', 'Ropa', 'Hogar', 'Deportes', 'Libros']
regiones   = ['AMBA', 'Centro', 'NOA', 'NEA', 'Cuyo', 'Patagonia']

df = pd.DataFrame({
    'fecha':      np.random.choice(fechas, n),
    'categoria':  np.random.choice(categorias, n, p=[0.3, 0.25, 0.2, 0.15, 0.1]),
    'region':     np.random.choice(regiones, n),
    'cantidad':   np.random.randint(1, 20, n),
    'precio':     np.round(np.random.uniform(500, 50000, n), 2),
    'devolucion': np.random.choice([True, False], n, p=[0.05, 0.95]),
})
df['ingreso'] = df['cantidad'] * df['precio'] * (~df['devolucion']).astype(int)
df['mes']     = pd.to_datetime(df['fecha']).dt.to_period('M')

print(f'Dataset: {df.shape[0]} registros, {df.shape[1]} columnas')
print(f'Período: {df["fecha"].min()} → {df["fecha"].max()}')
print(f'Ingreso total: ${df["ingreso"].sum():,.0f}')

# ─ Análisis 1: Ventas por categoría ─
print('\n=== Ventas por categoría ===')
por_cat = df.groupby('categoria').agg(
    registros  = ('ingreso', 'count'),
    ingreso    = ('ingreso', 'sum'),
    ticket_avg = ('precio',  'mean'),
).round(2).sort_values('ingreso', ascending=False)
print(por_cat)

# ─ Análisis 2: Tasa de devolución por región ─
print('\n=== Tasa de devolución por región ===')
dev_region = df.groupby('region')['devolucion'].agg(['sum','count'])
dev_region['tasa_%'] = (dev_region['sum'] / dev_region['count'] * 100).round(1)
print(dev_region.sort_values('tasa_%', ascending=False))

# ─ Análisis 3: Tendencia mensual con NumPy ─
print('\n=== Tendencia mensual de ingresos ===')
mensual = df.groupby('mes')['ingreso'].sum().reset_index()
mensual['mes_num'] = range(len(mensual))
# Ajuste lineal para ver tendencia
coef = np.polyfit(mensual['mes_num'], mensual['ingreso'], 1)
print(f'  Pendiente: ${coef[0]:+.0f} por mes (tendencia {"creciente ↑" if coef[0]>0 else "decreciente ↓"})')
for _, row in mensual.iterrows():
    barra = '█' * int(row['ingreso'] / mensual['ingreso'].max() * 20)
    print(f'  {str(row["mes"]):8} {barra:20} ${row["ingreso"]:>12,.0f}')

# ─ Análisis 4: Segmentación por precio (bins) ─
df['segmento'] = pd.cut(df['precio'],
                        bins=[0, 2000, 10000, 30000, 100000],
                        labels=['Económico','Medio','Premium','Lujo'])
print('\n=== Distribución por segmento de precio ===')
print(df['segmento'].value_counts().sort_index())
````
