## ⏱️ Bloque 1: Complejidad Algorítmica — Notación Big O

### 📘 Teoría

La **notación Big O** describe cómo crece el tiempo de ejecución (o el uso de memoria) de un algoritmo en función del tamaño de la entrada `n`.

#### Clases de complejidad (de mejor a peor)

| Notación | Nombre | Ejemplo |
|----------|--------|---------|
| O(1) | Constante | Acceder a `lista[5]` |
| O(log n) | Logarítmica | Búsqueda binaria |
| O(n) | Lineal | Recorrer una lista |
| O(n log n) | Lineal-logarítmica | Merge Sort, Quick Sort |
| O(n²) | Cuadrática | Bubble Sort, bucles anidados |
| O(2ⁿ) | Exponencial | Fibonacci recursivo sin memoización |

#### Reglas para calcular Big O
1. **Ignorar constantes:** `O(3n)` → `O(n)`
2. **Quedarse con el término dominante:** `O(n² + n)` → `O(n²)`
3. **Bucles anidados se multiplican:** dos `for` anidados sobre `n` → `O(n²)`
4. **Bucles seguidos se suman:** dos `for` separados → `O(n + n)` → `O(n)`

```python
# O(1) — tiempo constante
def primero(lista): return lista[0]

# O(n) — recorre toda la lista
def buscar(lista, objetivo):
    for elemento in lista:
        if elemento == objetivo: return True
    return False

# O(n²) — bucles anidados
def pares_duplicados(lista):
    for i in range(len(lista)):
        for j in range(i+1, len(lista)):
            if lista[i] == lista[j]: return True
    return False
```

### 💡 Ejemplo resuelto 1 — Medir y comparar tiempos reales

````python
import time
import random

# ✅ Comparar O(n) vs O(n²) con medición real

def tiempo_ejecucion(func, *args):
    t0 = time.perf_counter()
    result = func(*args)
    return time.perf_counter() - t0, result

# O(n) — suma de lista
def suma_lineal(lista):
    total = 0
    for x in lista:   # un solo recorrido
        total += x
    return total

# O(n²) — contar pares que suman objetivo
def pares_suma_objetivo(lista, objetivo):
    count = 0
    for i in range(len(lista)):
        for j in range(i+1, len(lista)):  # bucle anidado
            if lista[i] + lista[j] == objetivo:
                count += 1
    return count

# O(n) — mismo problema con set (hash lookup = O(1))
def pares_suma_objetivo_optimo(lista, objetivo):
    vistos = set()
    count = 0
    for x in lista:
        complemento = objetivo - x
        if complemento in vistos:
            count += 1
        vistos.add(x)
    return count

print(f'{'Tamaño':>10} | {'O(n) ms':>10} | {'O(n²) ms':>10} | {'O(n) opt ms':>12}')
print('-' * 52)

for n in [100, 500, 1000, 3000]:
    data = [random.randint(1, 100) for _ in range(n)]
    t1, _ = tiempo_ejecucion(suma_lineal, data)
    t2, r2 = tiempo_ejecucion(pares_suma_objetivo, data, 50)
    t3, r3 = tiempo_ejecucion(pares_suma_objetivo_optimo, data, 50)
    print(f'{n:>10} | {t1*1000:>10.3f} | {t2*1000:>10.3f} | {t3*1000:>12.3f}')

print('\n💡 Observá cómo O(n²) crece mucho más rápido que O(n)')
print('   Cuando n se duplica:')
print('   - O(n)  → tiempo se duplica')
print('   - O(n²) → tiempo se cuadruplica')
````

## 🔍 Bloque 2: Algoritmos de Búsqueda

### 📘 Teoría

#### Búsqueda Lineal — O(n)
Recorre la lista elemento por elemento hasta encontrar el objetivo.
```
Lista: [3, 7, 1, 9, 4, 6]   Buscar: 9
       ↑  ↑  ↑  ↑           → encontrado en posición 3
```
- **Ventaja:** funciona en listas desordenadas
- **Desventaja:** lento para listas grandes (O(n))

#### Búsqueda Binaria — O(log n)
Requiere lista **ordenada**. Divide el espacio de búsqueda por la mitad en cada paso.
```
Lista: [1, 3, 4, 6, 7, 9]   Buscar: 6
        ↑           ↑
        izq        der       medio = (0+5)//2 = 2 → lista[2]=4 < 6 → buscar derecha
                ↑   ↑
               izq der       medio = (3+5)//2 = 4 → lista[4]=7 > 6 → buscar izquierda
                ↑
              izq=der        medio = 3 → lista[3]=6 ✓ encontrado!
```
- **Ventaja:** muy rápido — en 1.000.000 de elementos, máximo 20 comparaciones
- **Desventaja:** requiere lista ordenada

**Comparación práctica:**
| n | Búsqueda lineal | Búsqueda binaria |
|---|----------------|------------------|
| 1.000 | 1.000 pasos | 10 pasos |
| 1.000.000 | 1.000.000 pasos | 20 pasos |
| 1.000.000.000 | 1.000.000.000 pasos | 30 pasos |

### 💡 Ejemplo resuelto 2 — Búsqueda lineal y binaria

````python
import time
import random

# ✅ Implementación y comparación de algoritmos de búsqueda

def busqueda_lineal(lista, objetivo):
    """O(n) — recorre la lista de izquierda a derecha."""
    for i, elemento in enumerate(lista):
        if elemento == objetivo:
            return i  # índice donde se encontró
    return -1  # no encontrado

def busqueda_binaria(lista, objetivo):
    """O(log n) — requiere lista ORDENADA."""
    izq, der = 0, len(lista) - 1

    while izq <= der:
        medio = (izq + der) // 2
        if lista[medio] == objetivo:
            return medio
        elif lista[medio] < objetivo:
            izq = medio + 1  # buscar en la mitad derecha
        else:
            der = medio - 1  # buscar en la mitad izquierda
    return -1

def busqueda_binaria_recursiva(lista, objetivo, izq=0, der=None):
    """Versión recursiva de la búsqueda binaria — O(log n)."""
    if der is None: der = len(lista) - 1
    if izq > der: return -1

    medio = (izq + der) // 2
    if lista[medio] == objetivo:   return medio
    elif lista[medio] < objetivo:  return busqueda_binaria_recursiva(lista, objetivo, medio+1, der)
    else:                          return busqueda_binaria_recursiva(lista, objetivo, izq, medio-1)

# Verificar correctitud
lista_ord = sorted(random.sample(range(1000), 50))
objetivo = lista_ord[20]  # elemento que sabemos que existe

idx_lin = busqueda_lineal(lista_ord, objetivo)
idx_bin = busqueda_binaria(lista_ord, objetivo)
idx_rec = busqueda_binaria_recursiva(lista_ord, objetivo)

print(f'Lista ordenada (primeros 10): {lista_ord[:10]}...')
print(f'Buscando: {objetivo}')
print(f'  Lineal:    índice {idx_lin} → valor {lista_ord[idx_lin]}')
print(f'  Binaria:   índice {idx_bin} → valor {lista_ord[idx_bin]}')
print(f'  Recursiva: índice {idx_rec} → valor {lista_ord[idx_rec]}')
print(f'  ¿Todos iguales? {idx_lin == idx_bin == idx_rec}')

# Comparar rendimiento
print('\n=== Comparación de rendimiento ===')
print(f'{'n':>10} | {'Lineal (ms)':>12} | {'Binaria (ms)':>13} | {'Speedup':>8}')
print('-' * 52)

for n in [1_000, 10_000, 100_000, 1_000_000]:
    datos = list(range(n))  # ya ordenado
    objetivo = n - 1        # peor caso: está al final

    t0 = time.perf_counter()
    for _ in range(100): busqueda_lineal(datos, objetivo)
    t_lin = (time.perf_counter() - t0) / 100

    t0 = time.perf_counter()
    for _ in range(100): busqueda_binaria(datos, objetivo)
    t_bin = (time.perf_counter() - t0) / 100

    speedup = t_lin / t_bin if t_bin > 0 else float('inf')
    print(f'{n:>10} | {t_lin*1000:>12.4f} | {t_bin*1000:>13.6f} | {speedup:>7.0f}x')
````

## 📊 Bloque 3: Algoritmos de Ordenamiento

### 📘 Teoría

| Algoritmo | Mejor caso | Caso promedio | Peor caso | Estable | Método |
|-----------|-----------|---------------|-----------|---------|--------|
| Bubble Sort | O(n) | O(n²) | O(n²) | ✅ | Comparación |
| Selection Sort | O(n²) | O(n²) | O(n²) | ❌ | Selección |
| Insertion Sort | O(n) | O(n²) | O(n²) | ✅ | Inserción |
| Merge Sort | O(n log n) | O(n log n) | O(n log n) | ✅ | División |
| Quick Sort | O(n log n) | O(n log n) | O(n²) | ❌ | Partición |

#### ¿Cuándo usar cada uno?
- **Bubble/Insertion:** para listas casi ordenadas o muy pequeñas
- **Merge Sort:** cuando necesitás O(n log n) garantizado y datos estables
- **Quick Sort:** el más rápido en la práctica para datos aleatorios
- **Python `sorted()`:** usa Timsort (híbrido Insertion + Merge) — O(n log n) garantizado

### 💡 Ejemplo resuelto 3 — Implementar y comparar algoritmos de ordenamiento

````python
import time
import random

# ✅ Implementación de los 5 algoritmos clásicos

def bubble_sort(arr):
    """O(n²) — compara pares adyacentes y los intercambia."""
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        intercambio = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                intercambio = True
        if not intercambio: break  # optimización: ya está ordenado
    return arr

def selection_sort(arr):
    """O(n²) — selecciona el mínimo y lo pone en su lugar."""
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        idx_min = i
        for j in range(i+1, n):
            if arr[j] < arr[idx_min]:
                idx_min = j
        arr[i], arr[idx_min] = arr[idx_min], arr[i]
    return arr

def insertion_sort(arr):
    """O(n²) — inserta cada elemento en su posición correcta."""
    arr = arr.copy()
    for i in range(1, len(arr)):
        clave = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > clave:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = clave
    return arr

def merge_sort(arr):
    """O(n log n) — divide y conquista: divide, ordena, combina."""
    if len(arr) <= 1: return arr
    medio = len(arr) // 2
    izq   = merge_sort(arr[:medio])
    der   = merge_sort(arr[medio:])
    return _merge(izq, der)

def _merge(izq, der):
    resultado, i, j = [], 0, 0
    while i < len(izq) and j < len(der):
        if izq[i] <= der[j]: resultado.append(izq[i]); i += 1
        else:                 resultado.append(der[j]); j += 1
    return resultado + izq[i:] + der[j:]

def quick_sort(arr):
    """O(n log n) promedio — elige pivote y particiona."""
    if len(arr) <= 1: return arr
    pivote  = arr[len(arr) // 2]
    menores = [x for x in arr if x < pivote]
    iguales = [x for x in arr if x == pivote]
    mayores = [x for x in arr if x > pivote]
    return quick_sort(menores) + iguales + quick_sort(mayores)

# Verificar correctitud
test = [64, 34, 25, 12, 22, 11, 90]
esperado = sorted(test)
algos = [bubble_sort, selection_sort, insertion_sort, merge_sort, quick_sort]
print('=== Verificación de correctitud ===')
for algo in algos:
    resultado = algo(test)
    ok = '✅' if resultado == esperado else '❌'
    print(f'  {ok} {algo.__name__:20}: {resultado}')

# Comparar rendimiento
print('\n=== Comparación de rendimiento (ms) ===')
print(f'{'n':>8} | {'bubble':>8} | {'select':>8} | {'insert':>8} | {'merge':>8} | {'quick':>8} | {'sorted':>8}')
print('-' * 70)

for n in [100, 500, 1000, 3000]:
    data = [random.randint(0, 10000) for _ in range(n)]
    tiempos = []
    for algo in algos + [sorted]:
        t0 = time.perf_counter()
        algo(data)
        tiempos.append((time.perf_counter()-t0)*1000)
    print(f'{n:>8} | ' + ' | '.join(f'{t:>8.2f}' for t in tiempos))
````

### 🚀 Ejercicio libre 3 — Merge Sort con conteo de inversiones

Una **inversión** es un par `(i, j)` donde `i < j` pero `lista[i] > lista[j]`.  
El número de inversiones mide cuán desordenada está una lista.

Modificá el algoritmo Merge Sort para que, además de ordenar, **cuente el número de inversiones** durante el proceso de merge.

Ejemplos:
- `[1, 2, 3, 4]` → 0 inversiones (ya ordenado)
- `[4, 3, 2, 1]` → 6 inversiones (completamente invertido)

## 🏗️ Bloque 4: Estructuras de Datos

### 📘 Teoría

#### Pila (Stack) — LIFO (Last In, First Out)
```
Apilar (push): [1] → [1,2] → [1,2,3]
Desapilar (pop):       [1,2,3] → [1,2] → [1]  (siempre el último)
```
Usos: historial de navegación, deshacer (Ctrl+Z), llamadas a funciones recursivas.

#### Cola (Queue) — FIFO (First In, First Out)
```
Encolar (enqueue): [1] → [1,2] → [1,2,3]
Desencolar (dequeue):     [1,2,3] → [2,3] → [3]  (siempre el primero)
```
Usos: cola de impresión, procesamiento de tareas, BFS en grafos.

#### Lista Enlazada (Linked List)
```
Nodo1 → Nodo2 → Nodo3 → None
  3       7       1
```
Cada nodo guarda un valor y una referencia al siguiente nodo.
- **Ventaja:** inserción/eliminación O(1) al principio
- **Desventaja:** acceso por índice O(n)

### 💡 Ejemplo resuelto 4 — Pila, Cola y Lista Enlazada desde cero

````python
# ✅ Implementación de las 3 estructuras de datos fundamentales

# ─────────────────────────────
# PILA (Stack)
# ─────────────────────────────
class Pila:
    """LIFO: el último en entrar es el primero en salir."""
    def __init__(self):
        self._datos = []

    def push(self, valor):     self._datos.append(valor)
    def pop(self):
        if self.esta_vacia(): raise IndexError('Pila vacía')
        return self._datos.pop()
    def peek(self):
        if self.esta_vacia(): raise IndexError('Pila vacía')
        return self._datos[-1]
    def esta_vacia(self):      return len(self._datos) == 0
    def tamaño(self):          return len(self._datos)
    def __repr__(self):        return f'Pila({self._datos})'

# Aplicación: verificar paréntesis balanceados
def parentesis_balanceados(expresion):
    """Usa una pila para verificar si los paréntesis están balanceados."""
    pila = Pila()
    pares = {')': '(', ']': '[', '}': '{'}
    for char in expresion:
        if char in '([{':
            pila.push(char)
        elif char in ')]}':
            if pila.esta_vacia() or pila.pop() != pares[char]:
                return False
    return pila.esta_vacia()

print('=== Pila ===')
p = Pila()
for v in [1, 2, 3, 4, 5]: p.push(v)
print(f'  Después de push 1-5: {p}')
print(f'  Peek: {p.peek()}')
print(f'  Pop: {p.pop()}, {p.pop()}')
print(f'  Estado final: {p}')

print('\n=== Paréntesis balanceados ===')
for expr in ['(a + b) * [c - d]', '((a + b)', '{[()]}', '([)]']:
    ok = parentesis_balanceados(expr)
    print(f'  {expr:20} → {'✅ Balanceado' if ok else '❌ No balanceado'}')

# ─────────────────────────────
# COLA (Queue)
# ─────────────────────────────
from collections import deque

class Cola:
    """FIFO: el primero en entrar es el primero en salir."""
    def __init__(self):
        self._datos = deque()  # deque es O(1) para ambos extremos

    def enqueue(self, valor):  self._datos.append(valor)
    def dequeue(self):
        if self.esta_vacia(): raise IndexError('Cola vacía')
        return self._datos.popleft()
    def frente(self):          return self._datos[0] if not self.esta_vacia() else None
    def esta_vacia(self):      return len(self._datos) == 0
    def tamaño(self):          return len(self._datos)
    def __repr__(self):        return f'Cola({list(self._datos)})'

print('\n=== Cola ===')
c = Cola()
for tarea in ['tarea_1', 'tarea_2', 'tarea_3']: c.enqueue(tarea)
print(f'  Cola inicial: {c}')
print(f'  Procesar: {c.dequeue()}')
print(f'  Procesar: {c.dequeue()}')
c.enqueue('tarea_4')
print(f'  Después de agregar tarea_4: {c}')

# ─────────────────────────────
# LISTA ENLAZADA
# ─────────────────────────────
class Nodo:
    def __init__(self, valor):
        self.valor    = valor
        self.siguiente = None

class ListaEnlazada:
    """Lista enlazada simple con inserción O(1) al inicio."""
    def __init__(self):
        self.cabeza = None
        self._tamaño = 0

    def insertar_inicio(self, valor):
        nuevo = Nodo(valor)
        nuevo.siguiente = self.cabeza
        self.cabeza = nuevo
        self._tamaño += 1

    def insertar_fin(self, valor):
        nuevo = Nodo(valor)
        if not self.cabeza:
            self.cabeza = nuevo
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo
        self._tamaño += 1

    def buscar(self, valor):
        actual = self.cabeza
        pos = 0
        while actual:
            if actual.valor == valor: return pos
            actual = actual.siguiente
            pos += 1
        return -1

    def eliminar(self, valor):
        if not self.cabeza: return False
        if self.cabeza.valor == valor:
            self.cabeza = self.cabeza.siguiente
            self._tamaño -= 1
            return True
        actual = self.cabeza
        while actual.siguiente:
            if actual.siguiente.valor == valor:
                actual.siguiente = actual.siguiente.siguiente
                self._tamaño -= 1
                return True
            actual = actual.siguiente
        return False

    def a_lista(self):
        resultado, actual = [], self.cabeza
        while actual:
            resultado.append(actual.valor)
            actual = actual.siguiente
        return resultado

    def __repr__(self): return ' → '.join(str(v) for v in self.a_lista()) + ' → None'

print('\n=== Lista Enlazada ===')
ll = ListaEnlazada()
for v in [1, 2, 3]: ll.insertar_fin(v)
ll.insertar_inicio(0)
print(f'  Lista: {ll}')
print(f'  Buscar 2: posición {ll.buscar(2)}')
ll.eliminar(2)
print(f'  Después de eliminar 2: {ll}')
````

## 🗺️ Bloque 5: Grafos y Árboles — BFS y DFS

### 📘 Teoría

Un **grafo** es un conjunto de nodos (vértices) conectados por aristas (edges).

```
    A ── B ── E
    │    │
    C ── D
```

#### Representaciones en código
```python
# Lista de adyacencia (la más común)
grafo = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'D'],
    'D': ['B', 'C'],
    'E': ['B']
}
```

#### BFS vs DFS

| | BFS (Breadth-First Search) | DFS (Depth-First Search) |
|---|---|---|
| **Estructura** | Cola (FIFO) | Pila (LIFO) o recursión |
| **Explora** | Nivel por nivel | Rama por rama |
| **Encuentra** | Camino más corto (sin pesos) | Cualquier camino |
| **Memoria** | Puede ser mayor (guarda todo el nivel) | Menor generalmente |
| **Usos** | GPS, redes sociales (grados de separación) | Laberintos, ciclos, topología |

### 💡 Ejemplo resuelto 5 — BFS y DFS implementados

````python
from collections import deque

# ✅ BFS y DFS sobre un grafo no dirigido

# Grafo que representa conexiones entre ciudades
grafo_ciudades = {
    'Buenos Aires': ['Rosario', 'Mar del Plata', 'La Plata'],
    'Rosario':      ['Buenos Aires', 'Córdoba', 'Santa Fe'],
    'Córdoba':      ['Rosario', 'Mendoza', 'Tucumán'],
    'Mar del Plata':['Buenos Aires'],
    'La Plata':     ['Buenos Aires'],
    'Santa Fe':     ['Rosario'],
    'Mendoza':      ['Córdoba', 'San Juan'],
    'Tucumán':      ['Córdoba', 'Salta'],
    'San Juan':     ['Mendoza'],
    'Salta':        ['Tucumán'],
}

def bfs(grafo, inicio):
    """BFS — explora nivel por nivel usando una Cola."""
    visitados = []
    cola = deque([inicio])
    visto = {inicio}

    while cola:
        nodo = cola.popleft()
        visitados.append(nodo)
        for vecino in grafo.get(nodo, []):
            if vecino not in visto:
                visto.add(vecino)
                cola.append(vecino)
    return visitados

def dfs(grafo, inicio, visitados=None):
    """DFS recursivo — explora rama por rama."""
    if visitados is None: visitados = []
    visitados.append(inicio)
    for vecino in grafo.get(inicio, []):
        if vecino not in visitados:
            dfs(grafo, vecino, visitados)
    return visitados

def camino_mas_corto_bfs(grafo, inicio, fin):
    """BFS para encontrar el camino más corto entre dos nodos."""
    cola = deque([[inicio]])
    visitados = {inicio}

    while cola:
        camino = cola.popleft()
        nodo = camino[-1]
        if nodo == fin:
            return camino
        for vecino in grafo.get(nodo, []):
            if vecino not in visitados:
                visitados.add(vecino)
                cola.append(camino + [vecino])
    return None  # no hay camino

print('=== BFS desde Buenos Aires ===')
orden_bfs = bfs(grafo_ciudades, 'Buenos Aires')
for i, ciudad in enumerate(orden_bfs):
    print(f'  Paso {i+1:2}: {ciudad}')

print('\n=== DFS desde Buenos Aires ===')
orden_dfs = dfs(grafo_ciudades, 'Buenos Aires')
for i, ciudad in enumerate(orden_dfs):
    print(f'  Paso {i+1:2}: {ciudad}')

print('\n=== Camino más corto ===')
pares = [('Buenos Aires', 'Salta'), ('Rosario', 'San Juan'), ('Mar del Plata', 'Tucumán')]
for origen, destino in pares:
    camino = camino_mas_corto_bfs(grafo_ciudades, origen, destino)
    print(f'  {origen} → {destino}:')
    print(f'    {' → '.join(camino)} ({len(camino)-1} saltos)')
````
