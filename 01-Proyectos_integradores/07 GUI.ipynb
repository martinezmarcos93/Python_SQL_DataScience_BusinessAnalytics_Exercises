## 📖 Conceptos base de GUIs

### ¿Qué es una GUI?
Una **GUI (Graphical User Interface)** es una interfaz que permite al usuario interactuar con el programa a través de elementos visuales en vez de texto.

### Componentes clave de cualquier GUI

| Componente | Descripción | PyQt5 | Tkinter |
|------------|-------------|-------|--------|
| **Ventana** | Contenedor principal | `QMainWindow` | `Tk()` |
| **Botón** | Dispara una acción | `QPushButton` | `Button` |
| **Etiqueta** | Muestra texto | `QLabel` | `Label` |
| **Campo de texto** | El usuario escribe | `QLineEdit` | `Entry` |
| **Lista** | Muestra ítems | `QListWidget` | `Listbox` |
| **Layout** | Organiza los widgets | `QVBoxLayout` | `pack/grid/place` |
| **Evento** | Responde a acciones | `clicked.connect()` | `command=función` |

### El flujo de una app GUI
```
1. Crear la ventana principal
2. Agregar widgets (botones, campos, etc.)
3. Conectar eventos a funciones (ej: clic → ejecutar cálculo)
4. Definir el layout (cómo se organizan los widgets)
5. Iniciar el loop de eventos (la app queda "escuchando")
```

### PyQt5 vs Tkinter

| | PyQt5 | Tkinter |
|--|-------|--------|
| **Instalación** | `pip install PyQt5` | Incluido en Python |
| **Aspecto** | Nativo del SO (se ve profesional) | Básico |
| **Complejidad** | Mayor (más features) | Menor (más simple) |
| **Uso ideal** | Apps de escritorio profesionales | Prototipado rápido, apps simples |
| **Licencia** | GPL / Comercial | Libre (parte de Python) |

````python
# ✅ SOLUCIÓN COMPLETA — Calculadora PyQt5
# Este código está listo para guardar como 'calculadora_pyqt.py' y ejecutar

codigo_calculadora = '''
import sys
from PyQt5.QtWidgets import (
    QApplication, QMainWindow, QWidget,
    QVBoxLayout, QGridLayout, QPushButton,
    QLineEdit, QLabel, QFrame
)
from PyQt5.QtCore import Qt
from PyQt5.QtGui import QFont, QPalette, QColor


class Calculadora(QMainWindow):
    """
    Calculadora de escritorio con PyQt5.
    Soporta: suma, resta, multiplicación, división, punto decimal y C (clear).
    """

    # --- COLORES (estilo oscuro tipo calculadora profesional) ---
    COLOR_FONDO      = '#1e1e2e'
    COLOR_DISPLAY    = '#2a2a3e'
    COLOR_NUM        = '#3a3a5e'
    COLOR_NUM_HOVER  = '#4a4a7e'
    COLOR_OP         = '#e67e22'
    COLOR_OP_HOVER   = '#f39c12'
    COLOR_IGUAL      = '#27ae60'
    COLOR_IGUAL_HOV  = '#2ecc71'
    COLOR_CLEAR      = '#c0392b'
    COLOR_TEXT       = '#ffffff'

    ESTILO_BTN_BASE = """
        QPushButton {{
            background-color: {bg};
            color: white;
            border: none;
            border-radius: 8px;
            font-size: 20px;
            font-weight: bold;
            min-height: 60px;
        }}
        QPushButton:hover {{
            background-color: {hover};
        }}
        QPushButton:pressed {{
            background-color: {pressed};
        }}
    """

    def __init__(self):
        super().__init__()

        # --- Estado interno de la calculadora ---
        self.numero_en_display = '0'   # Lo que se muestra actualmente
        self.numero_guardado   = None  # Primer operando
        self.operacion         = None  # Operación pendiente (+, -, *, /)
        self.esperar_nuevo     = False # True = próximo dígito limpia el display
        self.historial         = []    # Historial de operaciones

        self.initUI()

    # ================================================================
    # CONSTRUCCIÓN DE LA INTERFAZ
    # ================================================================
    def initUI(self):
        self.setWindowTitle('Calculadora PyQt5')
        self.setFixedSize(320, 520)  # Tamaño fijo (no redimensionable)
        self.setStyleSheet(f'background-color: {self.COLOR_FONDO};')

        # Widget central y layout principal
        central    = QWidget()
        layout_v   = QVBoxLayout()
        layout_v.setSpacing(8)
        layout_v.setContentsMargins(12, 12, 12, 12)

        # --- DISPLAY ---
        self.lbl_operacion = QLabel('')  # Muestra la operación anterior
        self.lbl_operacion.setAlignment(Qt.AlignRight)
        self.lbl_operacion.setStyleSheet(f'color: #888; font-size: 13px;')

        self.display = QLineEdit('0')    # El display principal
        self.display.setReadOnly(True)   # El usuario no tipea directamente
        self.display.setAlignment(Qt.AlignRight)
        self.display.setFont(QFont('Courier New', 32, QFont.Bold))
        self.display.setStyleSheet(f"""
            QLineEdit {{
                background-color: {self.COLOR_DISPLAY};
                color: white;
                border: none;
                border-radius: 8px;
                padding: 8px 12px;
            }}
        """)

        layout_v.addWidget(self.lbl_operacion)
        layout_v.addWidget(self.display)

        # --- GRID DE BOTONES ---
        # Definición de botones: (texto, fila, col, colspan, tipo)
        # tipo: 'num', 'op', 'igual', 'clear', 'especial'
        botones = [
            ('C',   0, 0, 1, 'clear'),  ('±',  0, 1, 1, 'especial'),
            ('%',   0, 2, 1, 'especial'),('÷',  0, 3, 1, 'op'),
            ('7',   1, 0, 1, 'num'),    ('8',  1, 1, 1, 'num'),
            ('9',   1, 2, 1, 'num'),    ('×',  1, 3, 1, 'op'),
            ('4',   2, 0, 1, 'num'),    ('5',  2, 1, 1, 'num'),
            ('6',   2, 2, 1, 'num'),    ('−',  2, 3, 1, 'op'),
            ('1',   3, 0, 1, 'num'),    ('2',  3, 1, 1, 'num'),
            ('3',   3, 2, 1, 'num'),    ('+',  3, 3, 1, 'op'),
            ('0',   4, 0, 2, 'num'),    (',',  4, 2, 1, 'num'),
            ('=',   4, 3, 1, 'igual'),
        ]

        grid = QGridLayout()
        grid.setSpacing(8)

        colores = {
            'num':      (self.COLOR_NUM,   self.COLOR_NUM_HOVER,  '#2a2a4e'),
            'op':       (self.COLOR_OP,    self.COLOR_OP_HOVER,   '#d35400'),
            'igual':    (self.COLOR_IGUAL, self.COLOR_IGUAL_HOV,  '#1e8449'),
            'clear':    (self.COLOR_CLEAR, '#e74c3c',             '#922b21'),
            'especial': ('#5a5a7e',        '#6a6a9e',             '#4a4a6e'),
        }

        for texto, fila, col, colspan, tipo in botones:
            btn = QPushButton(texto)
            bg, hover, pressed = colores[tipo]
            btn.setStyleSheet(self.ESTILO_BTN_BASE.format(
                bg=bg, hover=hover, pressed=pressed
            ))
            # Conectamos el clic pasando el texto del botón como argumento
            btn.clicked.connect(lambda checked, t=texto: self.manejar_click(t))
            grid.addWidget(btn, fila, col, 1, colspan)

        layout_v.addLayout(grid)
        central.setLayout(layout_v)
        self.setCentralWidget(central)

    # ================================================================
    # LÓGICA DE LA CALCULADORA
    # ================================================================
    def manejar_click(self, texto):
        """Distribuidor central: decide qué hacer según el botón presionado."""
        if texto.isdigit() or texto == ',':
            self._ingresar_digito(texto)
        elif texto in ('÷', '×', '−', '+'):
            self._ingresar_operacion(texto)
        elif texto == '=':
            self._calcular()
        elif texto == 'C':
            self._limpiar()
        elif texto == '±':
            self._cambiar_signo()
        elif texto == '%':
            self._porcentaje()

    def _ingresar_digito(self, digito):
        """Agrega un dígito al número en construcción."""
        if self.esperar_nuevo:
            self.numero_en_display = '0'
            self.esperar_nuevo = False

        # Manejo del punto decimal
        if digito == ',' and '.' in self.numero_en_display:
            return  # No permitir dos puntos decimales
        if digito == ',':
            digito = '.'

        if self.numero_en_display == '0' and digito != '.':
            self.numero_en_display = digito
        else:
            if len(self.numero_en_display) < 12:  # Límite de dígitos
                self.numero_en_display += digito

        self._actualizar_display()

    def _ingresar_operacion(self, op):
        """Guarda el operando actual y la operación a realizar."""
        if self.numero_guardado is not None and not self.esperar_nuevo:
            self._calcular()  # Calcular resultado intermedio si hay op pendiente

        self.numero_guardado = float(self.numero_en_display)
        self.operacion       = op
        self.esperar_nuevo   = True
        self.lbl_operacion.setText(
            f"{self._formatear(self.numero_guardado)} {op}"
        )

    def _calcular(self):
        """Ejecuta la operación pendiente y muestra el resultado."""
        if self.numero_guardado is None or self.operacion is None:
            return

        a = self.numero_guardado
        b = float(self.numero_en_display)

        operaciones = {
            '+': lambda x, y: x + y,
            '−': lambda x, y: x - y,
            '×': lambda x, y: x * y,
            '÷': lambda x, y: x / y if y != 0 else None,
        }

        resultado = operaciones[self.operacion](a, b)

        if resultado is None:
            self.numero_en_display = 'Error'
            self.lbl_operacion.setText('División por cero')
        else:
            expresion = f"{self._formatear(a)} {self.operacion} {self._formatear(b)} ="
            self.lbl_operacion.setText(expresion)
            self.historial.append(f"{expresion} {self._formatear(resultado)}")
            # Si el resultado es entero, lo mostramos sin decimales
            self.numero_en_display = str(int(resultado)) if resultado == int(resultado) else str(resultado)

        self.numero_guardado = None
        self.operacion       = None
        self.esperar_nuevo   = True
        self._actualizar_display()

    def _limpiar(self):
        """Resetea la calculadora al estado inicial."""
        self.numero_en_display = '0'
        self.numero_guardado   = None
        self.operacion         = None
        self.esperar_nuevo     = False
        self.lbl_operacion.setText('')
        self._actualizar_display()

    def _cambiar_signo(self):
        """Invierte el signo del número en pantalla."""
        valor = float(self.numero_en_display)
        valor = -valor
        self.numero_en_display = str(int(valor)) if valor == int(valor) else str(valor)
        self._actualizar_display()

    def _porcentaje(self):
        """Convierte el número actual a su valor porcentual."""
        valor = float(self.numero_en_display) / 100
        self.numero_en_display = str(valor)
        self._actualizar_display()

    def _actualizar_display(self):
        """Actualiza el texto del display con el número actual."""
        texto = self.numero_en_display
        # Ajustamos el tamaño de fuente según la longitud
        if len(texto) > 10:
            self.display.setFont(QFont('Courier New', 22, QFont.Bold))
        else:
            self.display.setFont(QFont('Courier New', 32, QFont.Bold))
        self.display.setText(texto)

    def _formatear(self, numero):
        """Formatea un número para mostrarlo: sin decimales si es entero."""
        return str(int(numero)) if numero == int(numero) else str(round(numero, 8))


if __name__ == "__main__":
    app = QApplication(sys.argv)
    app.setStyle('Fusion')  # Estilo moderno multiplataforma
    ventana = Calculadora()
    ventana.show()
    sys.exit(app.exec_())
'''

# Guardamos el código en un archivo .py
with open('calculadora_pyqt.py', 'w', encoding='utf-8') as f:
    f.write(codigo_calculadora)

print('✅ Archivo guardado: calculadora_pyqt.py')
print('\nPara ejecutar:')
print('  python calculadora_pyqt.py')
print('\nEsto abrirá una ventana de escritorio con la calculadora.')
````

````python
# ✅ VERSIÓN SIN GUI — La lógica pura de la calculadora (corre en el notebook)
# Esta versión te permite probar la lógica sin necesitar una ventana

class CalculadoraLogica:
    """
    La lógica de la calculadora sin interfaz gráfica.
    Útil para entender el funcionamiento y hacer tests.
    """
    def __init__(self):
        self.display       = '0'
        self.guardado      = None
        self.operacion     = None
        self.esperar_nuevo = False

    def presionar(self, tecla):
        """Simula presionar una tecla de la calculadora."""
        if tecla.isdigit() or tecla == '.':
            if self.esperar_nuevo:
                self.display = '0'
                self.esperar_nuevo = False
            self.display = (self.display + tecla) if self.display != '0' else tecla

        elif tecla in '+-*/':
            if self.guardado is not None and not self.esperar_nuevo:
                self.presionar('=')
            self.guardado    = float(self.display)
            self.operacion   = tecla
            self.esperar_nuevo = True

        elif tecla == '=':
            if self.guardado is not None and self.operacion:
                b = float(self.display)
                ops = {'+': lambda a,b: a+b, '-': lambda a,b: a-b,
                       '*': lambda a,b: a*b, '/': lambda a,b: a/b if b!=0 else None}
                resultado = ops[self.operacion](self.guardado, b)
                self.display = str(int(resultado)) if resultado and resultado == int(resultado) else str(resultado)
                self.guardado  = None
                self.operacion = None
                self.esperar_nuevo = True

        elif tecla == 'C':
            self.__init__()

        return self.display

    def secuencia(self, teclas):
        """Presiona una secuencia de teclas y muestra cada paso."""
        print(f'  Secuencia: {" ".join(teclas)}')
        for t in teclas:
            resultado = self.presionar(t)
        print(f'  Resultado: {resultado}\n')
        self.presionar('C')  # Reset para siguiente prueba
        return resultado


# Probamos la lógica
calc = CalculadoraLogica()

print('🧮 Pruebas de la lógica de la calculadora:')
print()
calc.secuencia(['5', '+', '3', '='])           # 5 + 3 = 8
calc.secuencia(['1', '2', '*', '4', '='])      # 12 * 4 = 48
calc.secuencia(['1', '0', '/', '4', '='])      # 10 / 4 = 2.5
calc.secuencia(['9', '9', '-', '1', '='])      # 99 - 1 = 98
calc.secuencia(['5', '+', '3', '+', '2', '=']) # 5 + 3 + 2 = 10 (encadenado)
calc.secuencia(['7', '/', '0', '='])           # 7 / 0 = None (división por cero)
````

### 📖 Explicación conceptual: Arquitectura de Tkinter

Tkinter usa un sistema de **geometry managers** para posicionar widgets:

| Manager | Cómo funciona | Cuándo usar |
|---------|--------------|-------------|
| `pack()` | Apila widgets horizontal o verticalmente | Layouts simples |
| `grid()` | Coloca en fila y columna | Formularios, grillas |
| `place()` | Posición absoluta en píxeles | Layouts precisos |

La estructura base de Tkinter:
```python
import tkinter as tk
from tkinter import ttk  # Widgets con mejor aspecto

class MiApp:
    def __init__(self, root):
        self.root = root
        self.root.title('Mi App')
        self.root.geometry('400x300')  # 'anchoXalto'
        self.crear_widgets()
    
    def crear_widgets(self):
        btn = tk.Button(self.root, text='Click', command=self.accion)
        btn.pack(pady=10)
    
    def accion(self):
        print('¡Botón presionado!')

root = tk.Tk()
app  = MiApp(root)
root.mainloop()  # Loop de eventos (equivalente a app.exec_() en PyQt5)
```

class OrganizadorTareas:
    def __init__(self, root):
        self.root   = root
        self.tareas = []  # Lista de dicts: {'texto': str, 'completada': bool, 'var': BooleanVar}

    def agregar_tarea(self):

````python
# ✅ SOLUCIÓN COMPLETA — Organizador de Tareas Tkinter

codigo_tareas = '''
import tkinter as tk
from tkinter import ttk, messagebox, font as tkfont
from datetime import datetime
import json
import os


class OrganizadorTareas:
    """
    Organizador de tareas con Tkinter.
    Features: agregar, completar, eliminar, filtrar, prioridades,
              persistencia en JSON, estadísticas.
    """

    ARCHIVO_DATOS = 'tareas.json'

    # Colores
    C_FONDO     = '#f0f4f8'
    C_HEADER    = '#2c3e50'
    C_BOTON_ADD = '#27ae60'
    C_BOTON_DEL = '#e74c3c'
    C_BOTON_OK  = '#2980b9'
    C_COMPLETA  = '#95a5a6'
    C_ALTA      = '#e74c3c'
    C_MEDIA     = '#e67e22'
    C_BAJA      = '#27ae60'

    def __init__(self, root):
        self.root  = root
        self.root.title('📋 Organizador de Tareas')
        self.root.geometry('620x680')
        self.root.configure(bg=self.C_FONDO)
        self.root.resizable(True, True)

        # Estado de la aplicación
        self.tareas       = []      # Lista de dicts con los datos de cada tarea
        self.filtro_actual = 'Todas' # Filtro activo

        self.cargar_tareas()        # Carga tareas guardadas (si existen)
        self.crear_widgets()        # Construye la interfaz
        self.actualizar_lista()     # Renderiza las tareas

    # ================================================================
    # CONSTRUCCIÓN DE LA INTERFAZ
    # ================================================================
    def crear_widgets(self):
        """Crea y organiza todos los widgets de la ventana."""

        # --- HEADER ---
        header = tk.Frame(self.root, bg=self.C_HEADER, pady=15)
        header.pack(fill='x')
        tk.Label(
            header, text='📋 Organizador de Tareas',
            bg=self.C_HEADER, fg='white',
            font=('Helvetica', 18, 'bold')
        ).pack()

        # --- PANEL DE ENTRADA ---
        frame_entrada = tk.LabelFrame(
            self.root, text=' Nueva tarea ',
            bg=self.C_FONDO, fg='#2c3e50',
            font=('Helvetica', 10, 'bold'),
            padx=10, pady=8
        )
        frame_entrada.pack(fill='x', padx=15, pady=(10, 5))

        # Campo de texto
        self.var_entrada = tk.StringVar()
        self.entrada = tk.Entry(
            frame_entrada, textvariable=self.var_entrada,
            font=('Helvetica', 12), relief='flat',
            bg='white', fg='#2c3e50'
        )
        self.entrada.pack(side='left', fill='x', expand=True, ipady=6, padx=(0, 8))
        # Enter también agrega la tarea
        self.entrada.bind('<Return>', lambda e: self.agregar_tarea())

        # Selector de prioridad
        tk.Label(frame_entrada, text='Prioridad:', bg=self.C_FONDO,
                 font=('Helvetica', 9)).pack(side='left')
        self.var_prioridad = tk.StringVar(value='Media')
        combo = ttk.Combobox(
            frame_entrada, textvariable=self.var_prioridad,
            values=['Alta', 'Media', 'Baja'], width=7, state='readonly'
        )
        combo.pack(side='left', padx=4)

        # Botón agregar
        tk.Button(
            frame_entrada, text='+ Agregar',
            command=self.agregar_tarea,
            bg=self.C_BOTON_ADD, fg='white',
            font=('Helvetica', 10, 'bold'),
            relief='flat', padx=10, cursor='hand2'
        ).pack(side='left', ipady=4)

        # --- BARRA DE FILTROS ---
        frame_filtros = tk.Frame(self.root, bg=self.C_FONDO)
        frame_filtros.pack(fill='x', padx=15, pady=4)

        tk.Label(frame_filtros, text='Filtrar:', bg=self.C_FONDO,
                 font=('Helvetica', 9)).pack(side='left')

        self.btns_filtro = {}
        for filtro in ['Todas', 'Pendientes', 'Completadas']:
            btn = tk.Button(
                frame_filtros, text=filtro,
                command=lambda f=filtro: self.aplicar_filtro(f),
                relief='flat', padx=10, pady=2,
                font=('Helvetica', 9), cursor='hand2'
            )
            btn.pack(side='left', padx=2)
            self.btns_filtro[filtro] = btn
        self.aplicar_filtro('Todas')  # Activa el filtro inicial

        # Búsqueda en tiempo real
        self.var_busqueda = tk.StringVar()
        self.var_busqueda.trace('w', lambda *args: self.actualizar_lista())
        tk.Entry(
            frame_filtros, textvariable=self.var_busqueda,
            font=('Helvetica', 9), width=15,
            fg='gray'
        ).pack(side='right', padx=4)
        tk.Label(frame_filtros, text='🔍', bg=self.C_FONDO).pack(side='right')

        # --- LISTA DE TAREAS ---
        frame_lista = tk.Frame(self.root, bg=self.C_FONDO)
        frame_lista.pack(fill='both', expand=True, padx=15, pady=5)

        # Scrollbar
        scrollbar = tk.Scrollbar(frame_lista)
        scrollbar.pack(side='right', fill='y')

        self.listbox = tk.Listbox(
            frame_lista,
            yscrollcommand=scrollbar.set,
            font=('Helvetica', 11),
            selectmode='single',       # Una sola selección a la vez
            relief='flat',
            bg='white',
            selectbackground='#3498db',
            selectforeground='white',
            activestyle='none',
            bd=0
        )
        self.listbox.pack(fill='both', expand=True)
        scrollbar.config(command=self.listbox.yview)
        # Doble clic = marcar/desmarcar
        self.listbox.bind('<Double-Button-1>', lambda e: self.toggle_completada())

        # --- BARRA DE ACCIONES ---
        frame_acciones = tk.Frame(self.root, bg=self.C_FONDO)
        frame_acciones.pack(fill='x', padx=15, pady=(5, 0))

        tk.Button(
            frame_acciones, text='✓ Completar/Descompletar',
            command=self.toggle_completada,
            bg=self.C_BOTON_OK, fg='white',
            font=('Helvetica', 10), relief='flat',
            padx=10, cursor='hand2'
        ).pack(side='left', ipady=4, padx=(0, 5))

        tk.Button(
            frame_acciones, text='🗑 Eliminar',
            command=self.eliminar_tarea,
            bg=self.C_BOTON_DEL, fg='white',
            font=('Helvetica', 10), relief='flat',
            padx=10, cursor='hand2'
        ).pack(side='left', ipady=4)

        tk.Button(
            frame_acciones, text='🗑 Eliminar completadas',
            command=self.eliminar_completadas,
            bg='#7f8c8d', fg='white',
            font=('Helvetica', 10), relief='flat',
            padx=10, cursor='hand2'
        ).pack(side='left', padx=5, ipady=4)

        # --- BARRA DE ESTADO (estadísticas) ---
        self.var_status = tk.StringVar()
        tk.Label(
            self.root, textvariable=self.var_status,
            bg='#dfe6e9', fg='#2c3e50',
            font=('Helvetica', 9), anchor='w', padx=10, pady=5
        ).pack(fill='x', side='bottom')

    # ================================================================
    # LÓGICA DE LA APLICACIÓN
    # ================================================================
    def agregar_tarea(self):
        """Agrega una nueva tarea a la lista."""
        texto = self.var_entrada.get().strip()
        if not texto:
            messagebox.showwarning('Atención', 'Escribí una tarea primero.')
            return

        tarea = {
            'texto':      texto,
            'completada': False,
            'prioridad':  self.var_prioridad.get(),
            'fecha':      datetime.now().strftime('%d/%m/%Y %H:%M')
        }
        self.tareas.append(tarea)
        self.var_entrada.set('')   # Limpiar campo
        self.entrada.focus()       # Volver el foco al campo
        self.guardar_tareas()
        self.actualizar_lista()

    def toggle_completada(self):
        """Alterna el estado completada/pendiente de la tarea seleccionada."""
        sel = self.listbox.curselection()
        if not sel:
            messagebox.showinfo('Info', 'Seleccioná una tarea primero.')
            return
        idx = self._idx_real(sel[0])
        self.tareas[idx]['completada'] = not self.tareas[idx]['completada']
        self.guardar_tareas()
        self.actualizar_lista()

    def eliminar_tarea(self):
        """Elimina la tarea seleccionada con confirmación."""
        sel = self.listbox.curselection()
        if not sel:
            messagebox.showinfo('Info', 'Seleccioná una tarea primero.')
            return
        idx   = self._idx_real(sel[0])
        texto = self.tareas[idx]['texto']
        if messagebox.askyesno('Confirmar', f'¿Eliminar "{texto}"?'):
            self.tareas.pop(idx)
            self.guardar_tareas()
            self.actualizar_lista()

    def eliminar_completadas(self):
        """Elimina todas las tareas marcadas como completadas."""
        cant = sum(1 for t in self.tareas if t['completada'])
        if cant == 0:
            messagebox.showinfo('Info', 'No hay tareas completadas.')
            return
        if messagebox.askyesno('Confirmar', f'¿Eliminar {cant} tarea(s) completada(s)?'):
            self.tareas = [t for t in self.tareas if not t['completada']]
            self.guardar_tareas()
            self.actualizar_lista()

    def aplicar_filtro(self, filtro):
        """Aplica un filtro (Todas / Pendientes / Completadas)."""
        self.filtro_actual = filtro
        # Resalta el botón activo
        for nombre, btn in self.btns_filtro.items():
            btn.config(
                bg='#2c3e50' if nombre == filtro else '#bdc3c7',
                fg='white'
            )
        self.actualizar_lista()

    def actualizar_lista(self):
        """Actualiza el Listbox aplicando el filtro y la búsqueda activos."""
        self.listbox.delete(0, tk.END)  # Limpia el listbox

        busqueda = self.var_busqueda.get().lower() if hasattr(self, 'var_busqueda') else ''

        colores_prioridad = {'Alta': self.C_ALTA, 'Media': self.C_MEDIA, 'Baja': self.C_BAJA}

        for tarea in self.tareas:
            # Aplicar filtro
            if self.filtro_actual == 'Pendientes' and tarea['completada']:
                continue
            if self.filtro_actual == 'Completadas' and not tarea['completada']:
                continue
            # Aplicar búsqueda
            if busqueda and busqueda not in tarea['texto'].lower():
                continue

            icono     = '✓' if tarea['completada'] else '○'
            prioridad = f"[{tarea['prioridad'][0]}]"  # [A], [M], [B]
            texto_item = f"  {icono} {prioridad} {tarea['texto']}"

            self.listbox.insert(tk.END, texto_item)

            idx = self.listbox.size() - 1
            if tarea['completada']:
                self.listbox.itemconfig(idx, fg=self.C_COMPLETA)
            else:
                self.listbox.itemconfig(idx, fg=colores_prioridad.get(tarea['prioridad'], '#2c3e50'))

        # Actualizar barra de estado
        total     = len(self.tareas)
        completadas = sum(1 for t in self.tareas if t['completada'])
        pendientes  = total - completadas
        pct = int(completadas / total * 100) if total > 0 else 0
        self.var_status.set(
            f"  Total: {total}  |  Pendientes: {pendientes}  |  Completadas: {completadas}  |  Progreso: {pct}%"
        )

    def _idx_real(self, idx_visual):
        """Mapea el índice visual del listbox al índice real en self.tareas."""
        busqueda = self.var_busqueda.get().lower() if hasattr(self, 'var_busqueda') else ''
        contador = 0
        for i, tarea in enumerate(self.tareas):
            if self.filtro_actual == 'Pendientes' and tarea['completada']:
                continue
            if self.filtro_actual == 'Completadas' and not tarea['completada']:
                continue
            if busqueda and busqueda not in tarea['texto'].lower():
                continue
            if contador == idx_visual:
                return i
            contador += 1
        return -1

    # ================================================================
    # PERSISTENCIA EN DISCO
    # ================================================================
    def guardar_tareas(self):
        """Guarda las tareas en un archivo JSON."""
        with open(self.ARCHIVO_DATOS, 'w', encoding='utf-8') as f:
            json.dump(self.tareas, f, ensure_ascii=False, indent=2)

    def cargar_tareas(self):
        """Carga las tareas desde el archivo JSON (si existe)."""
        if os.path.exists(self.ARCHIVO_DATOS):
            try:
                with open(self.ARCHIVO_DATOS, 'r', encoding='utf-8') as f:
                    self.tareas = json.load(f)
            except (json.JSONDecodeError, KeyError):
                self.tareas = []  # Si el archivo está corrupto, empezamos de cero


if __name__ == "__main__":
    root = tk.Tk()
    app  = OrganizadorTareas(root)
    root.mainloop()
'''

with open('organizador_tkinter.py', 'w', encoding='utf-8') as f:
    f.write(codigo_tareas)

print('✅ Archivo guardado: organizador_tkinter.py')
print('\nPara ejecutar:')
print('  python organizador_tkinter.py')
print('\nFeatures implementadas:')
features = [
    'Agregar tareas con Enter o botón',
    'Prioridades: Alta (rojo), Media (naranja), Baja (verde)',
    'Marcar/desmarcar como completada (clic en botón o doble clic)',
    'Eliminar tarea individual (con confirmación)',
    'Eliminar todas las completadas de una vez',
    'Filtros: Todas / Pendientes / Completadas',
    'Búsqueda en tiempo real',
    'Barra de estadísticas (total, pendientes, progreso %)',
    'Persistencia: las tareas se guardan en tareas.json',
]
for f in features:
    print(f'  ✓ {f}')
````

````python
# ✅ VERSIÓN SIN GUI — La lógica del organizador (corre en el notebook)

from datetime import datetime

class OrganizadorLogica:
    """Lógica del organizador de tareas sin interfaz gráfica."""

    def __init__(self):
        self.tareas = []

    def agregar(self, texto, prioridad='Media'):
        self.tareas.append({
            'texto': texto, 'completada': False,
            'prioridad': prioridad,
            'fecha': datetime.now().strftime('%d/%m %H:%M')
        })
        return self

    def completar(self, idx):
        if 0 <= idx < len(self.tareas):
            self.tareas[idx]['completada'] = not self.tareas[idx]['completada']
        return self

    def eliminar(self, idx):
        if 0 <= idx < len(self.tareas):
            self.tareas.pop(idx)
        return self

    def listar(self, filtro='Todas'):
        icono_prioridad = {'Alta': '🔴', 'Media': '🟡', 'Baja': '🟢'}
        print(f'\n=== TAREAS ({filtro}) ===')
        mostradas = 0
        for i, t in enumerate(self.tareas):
            if filtro == 'Pendientes'  and t['completada']:  continue
            if filtro == 'Completadas' and not t['completada']: continue
            estado = '✅' if t['completada'] else '⬜'
            prio   = icono_prioridad.get(t['prioridad'], '⚪')
            tachado = '~~' if t['completada'] else ''
            print(f'  [{i}] {estado} {prio} {tachado}{t["texto"]}{tachado}  ({t["fecha"]})')
            mostradas += 1
        if mostradas == 0:
            print('  (sin tareas)')
        total_comp = sum(1 for t in self.tareas if t['completada'])
        total      = len(self.tareas)
        pct        = int(total_comp/total*100) if total > 0 else 0
        print(f'  ─── {total_comp}/{total} completadas ({pct}%)')


# Demostración completa
org = OrganizadorLogica()

# Agregar tareas
org.agregar('Estudiar Python — módulo pandas', 'Alta')\
   .agregar('Completar ejercicio de SQL', 'Alta')\
   .agregar('Leer documentación de Statsmodels', 'Media')\
   .agregar('Practicar con el dataset de Iris', 'Media')\
   .agregar('Revisar apuntes de la semana', 'Baja')\
   .agregar('Instalar PyQt5 y probar la calculadora', 'Alta')

org.listar()                       # Ver todas
org.completar(0).completar(1)      # Completar las primeras dos
org.listar()                       # Ver todas con estado actualizado
org.listar('Pendientes')           # Solo pendientes
org.listar('Completadas')          # Solo completadas
org.eliminar(4)                    # Eliminar la tarea de índice 4
org.listar()                       # Estado final
````

### 🔑 Conceptos fundamentales de GUIs

| Concepto | Descripción |
|----------|-------------|
| **Widget** | Cualquier elemento visual: botón, campo, lista, etc. |
| **Layout** | Define cómo se organizan los widgets dentro del contenedor |
| **Señal/Evento** | Algo que ocurre: clic, tecla presionada, cambio de valor |
| **Slot/Callback** | Función que se ejecuta cuando ocurre el evento |
| **Loop de eventos** | Bucle que mantiene la ventana viva y procesa eventos |
| **Estado** | Variables internas que representan la situación actual de la app |
| **StringVar** | Variable reactiva de Tkinter que actualiza el widget automáticamente |
| **Pipeline** | Arquitectura: Vista (widgets) → Lógica (métodos) → Estado (variables) |
