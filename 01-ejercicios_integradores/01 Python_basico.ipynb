## 📦 Ejercicio 2 — Lista de números pares

**Enunciado:** Creá un programa que genere una lista de números pares del 1 al 20 usando un bucle `for` y luego imprima esa lista.

### 📖 Explicación conceptual

Un **bucle `for`** repite un bloque de código para cada elemento de una secuencia.

La función `range()` genera secuencias de números:

| Uso | Genera |
|-----|--------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(1, 10, 2)` | 1, 3, 5, 7, 9 (de 2 en 2) |

El operador **módulo** `%` devuelve el resto de una división: `10 % 2 == 0` significa que 10 es par.

Para agregar elementos a una lista usamos `.append()`:
```python
mi_lista = []
mi_lista.append(5)  # Ahora mi_lista es [5]
```

````python
# ✏️ TU CÓDIGO AQUÍ
# Creá una lista vacía y usá un for para agregar los números pares del 1 al 20
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

**Opción A (con if):** Iterá del 1 al 20, verificá si `numero % 2 == 0` y hacé `.append()`.

**Opción B (con range):** Usá `range(2, 21, 2)` para generar directamente los pares.

**Bonus:** ¿Podés hacerlo en una sola línea con list comprehension? `[x for x in range(...) if ...]`

</details>

````python
# ✅ SOLUCIÓN

# --- Método 1: usando for + if (más explícito, ideal para aprender) ---
pares = []
for numero in range(1, 21):  # range(1, 21) genera del 1 al 20
    if numero % 2 == 0:      # % es el módulo: devuelve el resto de la división
        pares.append(numero) # .append() agrega el elemento al final de la lista

print('Método 1 (for + if):', pares)

# --- Método 2: usando range con paso (más directo) ---
pares2 = list(range(2, 21, 2))  # range(inicio, fin, paso)
print('Método 2 (range con paso):', pares2)

# --- Método 3: list comprehension (estilo Pythónico) ---
pares3 = [x for x in range(1, 21) if x % 2 == 0]
print('Método 3 (list comprehension):', pares3)
````

**🔍 List Comprehension**  
Es una forma compacta de crear listas en Python. Sintaxis: `[expresión for elemento in iterable if condición]`. Es más rápida y considerada "pythónica" (código idiomático de Python).

## 📦 Ejercicio 3 — Filtrar nombres con 'A'

**Enunciado:** Dada una lista de nombres, creá un programa que filtre los nombres que comienzan con la letra 'A' y los almacene en una nueva lista.

### 📖 Explicación conceptual

Los strings en Python tienen **métodos** muy útiles:

| Método | Qué hace | Ejemplo |
|--------|----------|---------|
| `.startswith('A')` | True si empieza con 'A' | `'Ana'.startswith('A')` → True |
| `.upper()` | Convierte a mayúsculas | `'ana'.upper()` → 'ANA' |
| `.lower()` | Convierte a minúsculas | `'ANA'.lower()` → 'ana' |
| `.strip()` | Elimina espacios extra | `'  ana  '.strip()` → 'ana' |

También podés acceder a caracteres individuales con índices: `nombre[0]` es el primer carácter.

````python
# ✏️ TU CÓDIGO AQUÍ
nombres = ['Ana', 'Carlos', 'Alejandro', 'Beatriz', 'Alberto', 'Diana', 'Andrés', 'Elena']

# Filtrá los nombres que empiezan con 'A' y guardalos en una nueva lista
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

Iterá sobre `nombres` con un `for`. Para cada nombre, chequeá si `nombre.startswith('A')` o `nombre[0] == 'A'`.

¿Y si el usuario escribe 'alejandro' en minúscula? Usá `.upper()` o `.capitalize()` para normalizar antes de comparar.

</details>

````python
# ✅ SOLUCIÓN

nombres = ['Ana', 'Carlos', 'Alejandro', 'Beatriz', 'Alberto', 'Diana', 'Andrés', 'Elena']

# --- Método 1: for + if clásico ---
con_a = []
for nombre in nombres:
    if nombre.startswith('A'):  # .startswith() verifica el inicio del string
        con_a.append(nombre)

print('Nombres con A:', con_a)

# --- Método 2: list comprehension (más compacto) ---
con_a2 = [nombre for nombre in nombres if nombre.startswith('A')]
print('Nombres con A (comprehension):', con_a2)

# --- Método 3: robusto — funciona aunque estén en minúscula ---
con_a3 = [nombre for nombre in nombres if nombre.upper().startswith('A')]
print('Nombres con A (case-insensitive):', con_a3)

# Mostrar cuántos encontramos
print(f'\nSe encontraron {len(con_a)} nombres que empiezan con A.')
````

## 📦 Ejercicio 4 — Catálogo de productos

**Enunciado:** Creá un diccionario que represente un catálogo de productos (clave = nombre, valor = precio). Implementá una función que calcule el precio total de una lista de productos ingresada por el usuario.

### 📖 Explicación conceptual

Un **diccionario** es una colección de pares clave→valor:
```python
catalogo = {
    'manzana': 1.5,
    'pan': 2.0
}
```

| Operación | Código | Resultado |
|-----------|--------|-----------|
| Acceder | `catalogo['manzana']` | `1.5` |
| Agregar | `catalogo['leche'] = 3.0` | agrega un nuevo par |
| Verificar si existe | `'pan' in catalogo` | `True` |
| Acceso seguro | `catalogo.get('agua', 0)` | `0` (no lanza error) |
| Ver claves | `catalogo.keys()` | lista de nombres |

Una **función** se define con `def` y puede recibir parámetros y devolver valores con `return`.

````python
# ✏️ TU CÓDIGO AQUÍ
# 1. Definí el diccionario catalogo con al menos 5 productos
# 2. Definí una función calcular_total(productos_elegidos) que reciba una lista
#    de nombres y devuelva el precio total
# 3. Probá la función con una lista de ejemplo
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

Dentro de la función, iniciá una variable `total = 0`. Luego iterá sobre la lista de productos elegidos. Para cada uno, usá `catalogo.get(producto, 0)` para obtener el precio (si no existe, devuelve 0 en vez de dar error). Sumá al total y retornalo con `return`.

</details>

````python
# ✅ SOLUCIÓN

# Definimos el catálogo como diccionario {nombre: precio}
catalogo = {
    'manzana':  1.50,
    'banana':   0.80,
    'pan':      2.00,
    'leche':    3.50,
    'yogur':    4.00,
    'arroz':    2.80,
    'aceite':   6.50
}

def calcular_total(productos_elegidos):
    """
    Calcula el precio total de una lista de productos.
    
    Args:
        productos_elegidos (list): Lista de nombres de productos.
    
    Returns:
        float: El precio total.
    """
    total = 0
    no_encontrados = []
    
    for producto in productos_elegidos:
        precio = catalogo.get(producto, None)  # None si no existe
        if precio is not None:
            total += precio
        else:
            no_encontrados.append(producto)    # Registramos los que no encontramos
    
    if no_encontrados:
        print(f'⚠️  Productos no encontrados: {no_encontrados}')
    
    return total


# --- Probamos la función ---
mi_compra = ['manzana', 'leche', 'pan', 'café']  # 'café' no está en el catálogo

print('📦 Catálogo disponible:')
for producto, precio in catalogo.items():
    print(f'  {producto:<10} $ {precio:.2f}')   # :<10 alinea el texto

print(f'\n🛒 Tu compra: {mi_compra}')
total = calcular_total(mi_compra)
print(f'💰 Total a pagar: $ {total:.2f}')       # .2f = dos decimales
````

**🔍 Docstrings**  
El texto entre `"""triple comillas"""` dentro de una función se llama *docstring*. Documenta qué hace la función. Podés verlo con `help(calcular_total)`.

## 📦 Ejercicio 5 — Clase Persona

**Enunciado:** Definí una clase `Persona` con atributos `nombre`, `edad` y `género`. Creá un objeto e imprimí sus atributos.

### 📖 Explicación conceptual

La **Programación Orientada a Objetos (POO)** organiza el código en *clases* (plantillas) y *objetos* (instancias).

```
CLASE Persona  →  como un molde
OBJETO persona1 = Persona('Ana', 30, 'F')  →  como un producto del molde
```

| Concepto | Descripción |
|----------|-------------|
| `class` | Palabra clave para definir una clase |
| `__init__` | Método constructor: se ejecuta al crear el objeto |
| `self` | Referencia al objeto actual |
| Atributo | Variable que pertenece al objeto (`self.nombre`) |
| Método | Función que pertenece a la clase |

````python
# ✏️ TU CÓDIGO AQUÍ
# Definí la clase Persona con __init__ y un método presentarse()
# Creá al menos dos objetos y mostrá sus atributos
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

```python
class Persona:
    def __init__(self, nombre, edad, genero):
        self.nombre = nombre   # 'self.nombre' es el atributo del objeto
        # ... completá el resto
    
    def presentarse(self):
        print(f'Hola, soy {self.nombre}...')  # usá los atributos con self.
```
Después: `p1 = Persona('Ana', 30, 'F')` y `p1.presentarse()`

</details>

````python
# ✅ SOLUCIÓN

class Persona:
    """Representa una persona con nombre, edad y género."""
    
    def __init__(self, nombre, edad, genero):
        """Constructor: se ejecuta automáticamente al crear un objeto."""
        self.nombre = nombre   # self.atributo guarda el valor en el objeto
        self.edad   = edad
        self.genero = genero
    
    def presentarse(self):
        """Imprime una presentación de la persona."""
        print(f'Hola, me llamo {self.nombre}, tengo {self.edad} años y mi género es {self.genero}.')
    
    def es_mayor_de_edad(self):
        """Devuelve True si la persona tiene 18 años o más."""
        return self.edad >= 18
    
    def __str__(self):
        """Representación en texto del objeto (usado por print())."""
        return f'Persona({self.nombre}, {self.edad}, {self.genero})'


# Creamos objetos (instancias) de la clase
persona1 = Persona('Ana García', 28, 'Femenino')
persona2 = Persona('Carlos López', 15, 'Masculino')

# Accedemos a atributos y métodos
persona1.presentarse()
persona2.presentarse()

print(f'\n{persona1.nombre} es mayor de edad: {persona1.es_mayor_de_edad()}')
print(f'{persona2.nombre} es mayor de edad: {persona2.es_mayor_de_edad()}')

# __str__ se llama automáticamente cuando usamos print() con el objeto
print(f'\nRepresentación del objeto: {persona1}')
````

## 📦 Ejercicio 6 — Herencia: clase Estudiante

**Enunciado:** Extendé la clase `Persona` con una subclase `Estudiante` que tenga el atributo adicional `curso`. Mostrá atributos heredados y propios.

### 📖 Explicación conceptual

La **herencia** permite que una clase *hija* reutilice el código de una clase *padre*.

```
Persona (padre)  →  tiene nombre, edad, genero
    ↓ hereda
Estudiante (hijo) →  tiene TODO lo de Persona + curso, promedio
```

- `super()` llama al constructor del padre para no repetir código.
- El hijo puede **sobreescribir** métodos del padre (override).

````python
# ✏️ TU CÓDIGO AQUÍ
# Definí la clase Estudiante que herede de Persona
# Agregale los atributos: curso y promedio
# Sobreescribí el método presentarse() para incluir la info del estudiante
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

```python
class Estudiante(Persona):  # Entre paréntesis va la clase padre
    def __init__(self, nombre, edad, genero, curso, promedio):
        super().__init__(nombre, edad, genero)  # Llama al __init__ de Persona
        self.curso    = curso    # Atributos propios de Estudiante
        self.promedio = promedio
```

</details>

````python
# ✅ SOLUCIÓN

class Estudiante(Persona):  # Estudiante HEREDA de Persona
    """Representa un estudiante. Hereda de Persona y agrega curso y promedio."""
    
    def __init__(self, nombre, edad, genero, curso, promedio):
        # super() llama al __init__ de la clase padre (Persona)
        # Así no repetimos el código de nombre, edad, genero
        super().__init__(nombre, edad, genero)
        
        # Atributos nuevos propios de Estudiante
        self.curso    = curso
        self.promedio = promedio
    
    def presentarse(self):
        """Override: versión mejorada de presentarse para estudiantes."""
        # Llamamos al método del padre primero
        super().presentarse()
        # Y luego agregamos info propia
        print(f'  → Estoy cursando: {self.curso} | Promedio: {self.promedio:.1f}')
    
    def aprobado(self, minimo=6.0):
        """Devuelve True si el promedio supera el mínimo requerido."""
        return self.promedio >= minimo


# Creamos un Estudiante
est1 = Estudiante('María Torres', 22, 'Femenino', 'Ciencia de Datos', 8.5)
est2 = Estudiante('Pedro Ruiz',   19, 'Masculino', 'Programación',    5.3)

print('=== Presentaciones ===')
est1.presentarse()
print()
est2.presentarse()

print('\n=== Verificación de aprobados ===')
for est in [est1, est2]:
    estado = '✅ Aprobado' if est.aprobado() else '❌ Desaprobado'
    print(f'{est.nombre}: {estado}')

# Verificamos herencia
print(f'\n¿est1 es instancia de Estudiante? {isinstance(est1, Estudiante)}')
print(f'¿est1 es instancia de Persona?    {isinstance(est1, Persona)}')
````

## 📦 Ejercicio 7 — Manipulación de archivos (nombres y apellidos)

**Enunciado:** Creá un programa que lea un archivo de texto con nombres y apellidos, y genere un nuevo archivo con solo los apellidos.

### 📖 Explicación conceptual

En Python abrimos archivos con `open()`. Siempre es mejor usar el bloque `with` porque **cierra el archivo automáticamente** aunque haya errores.

| Modo | Descripción |
|------|-------------|
| `'r'` | Leer (read) |
| `'w'` | Escribir (write) — borra el contenido anterior |
| `'a'` | Agregar (append) — no borra el contenido anterior |

```python
with open('archivo.txt', 'r') as f:
    lineas = f.readlines()  # Lee todas las líneas en una lista
```

````python
# Primero creamos el archivo de ejemplo para el ejercicio
contenido_ejemplo = """Ana García
Carlos López
María Torres
Pedro Ramírez
Lucía Fernández
Juan Martínez
"""

with open('nombres_completos.txt', 'w', encoding='utf-8') as f:
    f.write(contenido_ejemplo)

print('Archivo de ejemplo creado: nombres_completos.txt')
print('Contenido:')
print(contenido_ejemplo)
````

````python
# ✏️ TU CÓDIGO AQUÍ
# 1. Abrí 'nombres_completos.txt' y leé las líneas
# 2. Para cada línea, quedate solo con el apellido (segunda palabra)
# 3. Guardá los apellidos en 'apellidos.txt'
# 4. Mostrá los apellidos encontrados
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

- `linea.strip()` elimina el salto de línea `\n` al final.
- `linea.split()` divide el texto en palabras → `['Ana', 'García']`
- El apellido está en la posición `[1]` (índice 1).
- Para escribir: `f.write(apellido + '\n')` agrega un salto de línea.

</details>

````python
# ✅ SOLUCIÓN

apellidos = []

# Leemos el archivo de entrada
with open('nombres_completos.txt', 'r', encoding='utf-8') as archivo_entrada:
    for linea in archivo_entrada:
        linea = linea.strip()          # Elimina espacios y saltos de línea (\n)
        if linea:                      # Ignoramos líneas vacías
            partes  = linea.split()    # Separa por espacios: ['Ana', 'García']
            apellido = partes[-1]      # [-1] toma el último elemento (apellido)
            apellidos.append(apellido)

# Escribimos el archivo de salida
with open('apellidos.txt', 'w', encoding='utf-8') as archivo_salida:
    for apellido in apellidos:
        archivo_salida.write(apellido + '\n')  # \n = salto de línea

# Mostramos el resultado
print(f'Se extrajeron {len(apellidos)} apellidos:')
for apellido in apellidos:
    print(f'  → {apellido}')

print('\n✅ Archivo "apellidos.txt" creado exitosamente.')
````

## 📦 Ejercicio 8 — Procesar CSV de ventas

**Enunciado:** Tomá un archivo CSV con datos de ventas (producto, cantidad, precio) y calculá el total de ventas por producto. Guardá los resultados en un nuevo CSV.

### 📖 Explicación conceptual

Un **CSV** (Comma-Separated Values) es un archivo de texto donde los datos se separan por comas. Python tiene el módulo `csv` para trabajar con ellos fácilmente.

```
producto,cantidad,precio
manzana,10,1.50
banana,5,0.80
```

El módulo `csv.DictReader` lee cada fila como un diccionario `{'producto': 'manzana', 'cantidad': '10', ...}`

````python
# Creamos el CSV de ejemplo
import csv

datos_ventas = [
    ['producto', 'cantidad', 'precio'],
    ['manzana', '10', '1.50'],
    ['banana',  '5',  '0.80'],
    ['manzana', '8',  '1.50'],
    ['pan',     '20', '2.00'],
    ['banana',  '12', '0.80'],
    ['pan',     '15', '2.00'],
    ['leche',   '7',  '3.50'],
]

with open('ventas.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(datos_ventas)

print('Archivo ventas.csv creado con', len(datos_ventas)-1, 'registros.')
````

````python
# ✏️ TU CÓDIGO AQUÍ
# 1. Leé ventas.csv con csv.DictReader
# 2. Calculá el total (cantidad * precio) por producto usando un diccionario
# 3. Guardá los resultados en 'totales_ventas.csv'
# 4. Mostrá los resultados por pantalla
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

- Creá un diccionario `totales = {}` para acumular.
- Para cada fila: `total_fila = int(fila['cantidad']) * float(fila['precio'])`.
- Para acumular: `totales[producto] = totales.get(producto, 0) + total_fila`.
- Para escribir el CSV de salida, usá `csv.DictWriter` con `fieldnames=['producto', 'total_ventas']`.

</details>

````python
# ✅ SOLUCIÓN
import csv

totales = {}  # Diccionario para acumular: {producto: total_ventas}

# Leemos el CSV de entrada
with open('ventas.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)   # Cada fila es un dict con las columnas como claves
    for fila in reader:
        producto  = fila['producto']
        cantidad  = int(fila['cantidad'])
        precio    = float(fila['precio'])
        subtotal  = cantidad * precio
        
        # .get(clave, valor_default) evita KeyError si el producto no existe aún
        totales[producto] = totales.get(producto, 0) + subtotal

# Mostramos los resultados
print('📊 Total de ventas por producto:')
print(f'{"Producto":<12} {"Total ($)":>10}')
print('-' * 24)
for producto, total in sorted(totales.items()):
    print(f'{producto:<12} {total:>10.2f}')

# Guardamos el CSV de salida
with open('totales_ventas.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['producto', 'total_ventas'])
    writer.writeheader()           # Escribe la fila de encabezados
    for producto, total in totales.items():
        writer.writerow({'producto': producto, 'total_ventas': round(total, 2)})

print('\n✅ Resultados guardados en "totales_ventas.csv"')
````

## 📦 Ejercicio 9 — NumPy: array y operaciones

**Enunciado:** Usá NumPy para crear un array con los números del 1 al 10, elevar todos los elementos al cuadrado y mostrar el resultado.

### 📖 Explicación conceptual

**NumPy** es la biblioteca fundamental para computación numérica en Python. Su ventaja principal: las operaciones se aplican a **todos los elementos a la vez** (vectorización), sin necesidad de bucles.

```python
import numpy as np

a = np.array([1, 2, 3])  # Array de NumPy
a * 2                     # → [2, 4, 6]  (¡sin for!)
a ** 2                    # → [1, 4, 9]
```

| Función | Descripción |
|---------|-------------|
| `np.array([...])` | Crea un array desde una lista |
| `np.arange(1, 11)` | Array del 1 al 10 |
| `np.zeros(5)` | Array de 5 ceros |
| `np.mean(a)` | Promedio |
| `np.sum(a)` | Suma total |

````python
# ✏️ TU CÓDIGO AQUÍ
# 1. Importá numpy como np
# 2. Creá un array del 1 al 10
# 3. Elevalo al cuadrado y mostrá el resultado
# 4. Calculá la suma y el promedio de los cuadrados
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

- `np.arange(1, 11)` genera `[1, 2, 3, ..., 10]`
- Para elevar al cuadrado: `array ** 2` o `np.power(array, 2)`
- Para la suma: `np.sum(cuadrados)` o `cuadrados.sum()`

</details>

````python
# ✅ SOLUCIÓN
import numpy as np

# Creamos el array del 1 al 10
numeros = np.arange(1, 11)  # arange(inicio, fin_exclusivo)
print(f'Array original:   {numeros}')

# Elevamos al cuadrado — NumPy aplica la operación a CADA elemento
cuadrados = numeros ** 2
print(f'Al cuadrado:      {cuadrados}')

# Operaciones estadísticas
print(f'\nSuma:             {np.sum(cuadrados)}')
print(f'Promedio:         {np.mean(cuadrados):.2f}')
print(f'Mínimo:           {np.min(cuadrados)}')
print(f'Máximo:           {np.max(cuadrados)}')
print(f'Desv. estándar:   {np.std(cuadrados):.2f}')

# Comparación de velocidad: NumPy vs Python puro
import time

N = 1_000_000
arr_np = np.arange(N)

t0 = time.time()
resultado_numpy = arr_np ** 2
t1 = time.time()
print(f'\nNumPy tardó:       {(t1-t0)*1000:.2f} ms para {N:,} elementos')

t0 = time.time()
resultado_python = [x**2 for x in range(N)]
t1 = time.time()
print(f'Python puro tardó: {(t1-t0)*1000:.2f} ms para {N:,} elementos')
````

## 📦 Ejercicio 10 — pandas: análisis básico de ventas

**Enunciado:** Usá pandas para cargar un CSV de ventas y calculá la media, mediana y desviación estándar de los precios.

### 📖 Explicación conceptual

**pandas** es la biblioteca estrella para análisis de datos. Trabaja con **DataFrames**: tablas con filas y columnas, como una hoja de cálculo en Python.

```python
import pandas as pd

df = pd.read_csv('ventas.csv')  # Lee un CSV → DataFrame
df.head()                        # Muestra las primeras 5 filas
df['precio'].mean()              # Promedio de la columna 'precio'
df.describe()                    # Resumen estadístico completo
```

````python
# ✏️ TU CÓDIGO AQUÍ
# 1. Importá pandas como pd
# 2. Cargá el archivo 'ventas.csv' creado anteriormente
# 3. Mostrá las primeras filas y la info general
# 4. Calculá media, mediana y desviación estándar del precio
# 5. Mostrá el total de ventas por producto (cantidad * precio)
````

<details>
<summary>💡 Pista (hacé clic para ver)</summary>

- `df['precio'].mean()` → media
- `df['precio'].median()` → mediana  
- `df['precio'].std()` → desviación estándar
- Para crear una nueva columna: `df['total'] = df['cantidad'] * df['precio']`
- Para agrupar: `df.groupby('producto')['total'].sum()`

</details>

````python
# ✅ SOLUCIÓN
import pandas as pd

# Cargamos el CSV en un DataFrame
df = pd.read_csv('ventas.csv')

print('=== Vista general del DataFrame ===')
print(df)

print('\n=== Información del DataFrame ===')
print(df.info())                # Tipos de datos y valores nulos

print('\n=== Estadísticas del precio ===')
print(f'Media:              $ {df["precio"].mean():.2f}')
print(f'Mediana:            $ {df["precio"].median():.2f}')
print(f'Desv. estándar:     $ {df["precio"].std():.2f}')
print(f'Mínimo:             $ {df["precio"].min():.2f}')
print(f'Máximo:             $ {df["precio"].max():.2f}')

print('\n=== Resumen estadístico completo (describe) ===')
print(df.describe().round(2))  # .describe() calcula todo de una vez

# Calculamos el total por fila y agrupamos por producto
df['total'] = df['cantidad'] * df['precio']

print('\n=== Total de ventas por producto ===')
ventas_por_producto = df.groupby('producto')['total'].sum().sort_values(ascending=False)
print(ventas_por_producto.round(2))

print(f'\n💰 GRAN TOTAL: $ {df["total"].sum():.2f}')
````
