```python
# Detección automática de anomalías

def detectar_anomalias_zscore(serie, umbral=2.5):
    """
    Detecta anomalías usando Z-score.
    Un punto es anomalía si su z-score > umbral.
    """
    media = serie.rolling(window=12).mean()
    std = serie.rolling(window=12).std()
    z_scores = np.abs((serie - media) / std)
    return z_scores > umbral

# TU CÓDIGO AQUÍ:
# 1. Aplicá la detección de anomalías a ventas_unidades
# 2. Graficá la serie con los puntos anómalos marcados en rojo
# 3. ¿Las semanas anómalas coinciden con eventos conocidos
#    (Black Friday, caídas del sistema, promociones)?
```
