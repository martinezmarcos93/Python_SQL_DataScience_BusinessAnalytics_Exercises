## 📌 PARTE 4: Producto

### Objetivos
- Calcular el costo total del producto
- Estimar el precio de venta y la ganancia
- Analizar estadísticas de rendimiento en el mercado

### 💡 Pistas
- Fórmula clave: `Ganancia = Precio_venta - Costo_total`
- `Margen (%) = (Ganancia / Precio_venta) * 100`
- Comparar con márgenes de la industria obtenidos en el análisis de benchmarks

````python
# Estructura de costos del producto

def calcular_estructura_costos(
    materia_prima,
    mano_de_obra,
    costos_indirectos,
    gastos_admin_pct=0.10,
    margen_deseado_pct=0.30
):
    """
    Calcula la estructura de costos y el precio sugerido de venta.
    
    Args:
        materia_prima: Costo de insumos por unidad
        mano_de_obra: Costo laboral por unidad
        costos_indirectos: Alquiler, servicios, etc. por unidad
        gastos_admin_pct: % de gastos administrativos sobre costo directo
        margen_deseado_pct: % de margen de ganancia deseado
    """
    costo_directo = materia_prima + mano_de_obra + costos_indirectos
    gastos_admin = costo_directo * gastos_admin_pct
    costo_total = costo_directo + gastos_admin
    precio_venta = costo_total / (1 - margen_deseado_pct)
    ganancia = precio_venta - costo_total
    
    return {
        'Costo_directo': costo_directo,
        'Gastos_admin': gastos_admin,
        'Costo_total': costo_total,
        'Precio_venta': precio_venta,
        'Ganancia': ganancia,
        'Margen_%': margen_deseado_pct * 100
    }

# TU CÓDIGO AQUÍ:
# 1. Llamá a la función con los valores de tu producto
# 2. Imprimí los resultados de forma clara
# 3. Comparar con el precio promedio de la competencia (Parte 1b)
# 4. ¿El precio es competitivo?
````

````python
# === RATIOS FINANCIEROS ===

def calcular_ratios_financieros(activo_corriente, pasivo_corriente, 
                                 ventas, costo_ventas, 
                                 ganancia_neta, patrimonio_neto,
                                 deuda_total, activo_total):
    ratios = {
        # Liquidez
        'Liquidez_corriente': activo_corriente / pasivo_corriente,  # >1.5 es bueno
        
        # Rentabilidad
        'Margen_bruto_%': ((ventas - costo_ventas) / ventas) * 100,
        'Margen_neto_%': (ganancia_neta / ventas) * 100,
        'ROE_%': (ganancia_neta / patrimonio_neto) * 100,  # Retorno sobre patrimonio
        'ROA_%': (ganancia_neta / activo_total) * 100,     # Retorno sobre activos
        
        # Endeudamiento
        'Razon_deuda': deuda_total / activo_total,  # <0.5 es conservador
    }
    return pd.Series(ratios)

# TU CÓDIGO AQUÍ:
# 1. Ingresá los valores financieros de tu startup
# 2. Llamá a la función y mostrá los resultados
# 3. Comparalos con los ratios típicos de la industria
````
